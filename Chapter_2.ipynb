##data relational
ฐานข้อมูลเชิงสัมพันธ์   คือการจัดเก็บรวบรวมข้อมูลในลักษณะตารางสองมิติซึ่งประกอบด้วยแถวและคอลัมน์ ซึ่งข้อมูลเหล่านี้จะมีความสัมพันธ์ระหว่างกัน

1)โครงสร้างของฐานข้อมูลเชิงสัมพันธ์
โครงสร้างของฐานข้อมูลเชิงสัมพันธ์   มีการใช้โครงสร้างข้อมูลในเชิงตรรกะเพียงรูปแบบเดียวเท่านั้น คือรีเลชัน (Relation)

โดยทีรีเลชันจะถูกมองเห็นในลักษณะของตาราง (Table) ที่มีคุณสมบัติดังต่อไปนี้

1.1) แต่ละแถวใช้แทนทัปเพิล (Tuple) หรือ เรคอร์ด (Record) ในรีเลชัน
1.2) ลำดับของทัปเพิลไม่มีความสำคัญ
1.3) ลำดับของคอลัมน์ไม่มีความสำคัญ
1.4) ทุกทัปเพิลต้องมีความแตกต่างกันโดยเนื้อหาหรือข้อเท็จจริง หรือกล่าวอีกอย่างหนึ่งก็คือ ข้อมูลในแต่ละทัปเพิลต้องไม่ซ้ำกันนั่นเอง ตัวอย่างของรีเลชันแสดงไว้ในภาพข้างล่าง

relation


2) เอนทิตีและรีเลชันชิพ (Entity and Relationship)

2.1) เอนทิตี (Entity)
หมายถึงวัตถุหรือสิ่งที่เราสนใจต้องการเก็บข้อมูล มีเอกลักษณ์ในตัวเอง  คือสามารถบอกความแตกต่างของแต่ละเอนทิตีได้ และจะถูกสร้างขึ้นโดยชุดของแอตทรบิวต์

2.2) เอนทิตีเซต (Entity Set)
คือ  ชุดของเอนทิตีชนิดเดียวกัน คุณสมบัติของเอนทิตีแต่ละตัวสามารถอธิบายโดยใช้ชุดของแอตทริบิวต์

2.3) รีเลชันชิพ (Relationship)
หมายถึงความสัมพันธ์ระหว่างเอนทิตี


3) แอตทริบิวต์ (Attributes) และโดเมน (Domains)
แอตทริบิวต์ หมายถึง คุณลักษณะเฉพาะของแต่ละเอนทิตีและค่าในแต่ละแอตทริบิวต์เป็นค่าที่นำมาจากโดเมนใดโดเมนหนึ่งเท่านั้น โดยที่โดเมน คือเซตของค่าที่เป็นไปได้ในแอตทริบิวต์นั้น

3.1) คุณสมบัติของค่าแอตทริบิวต์
ค่าของแอตทริบิวต์ที่จะใส่ลงไปในรีเลชันต้องมีคุณสมบัติดังต่อไปนี้

1. เป็นค่าที่อะตอมมิค (Atomic) หมายถึงเป็นค่าที่ไม่สามาถแบ่งย่อยต่อไปได้อีก

2. ไม่อนุญาตให้มีแอตทริบิวต์ที่มีหลายค่า (Multivalued Attribute) ซึ่งเป็นคุณสมบัติ
ของการทำเฟิสท์นอร์มัลฟอร์ม (First Normal Form)

3.2) คุณสมบัติของโดเมน
แต่ละโดเมนจะถูกกำหนดด้วย ชื่อ ชนิดของข้อมูล (Data Type)  และรูปแบบ(Format) ดัง
ตัวอย่างในภาพข้างล่าง

domain


3.3) ชื่อของคอลัมน์หรือชื่อของแอตทริบิวต์
คอลัมน์มีคุณสมบัติดังต่อไปนี้

1. แต่ละคอลัมน์มีการกำหนดชื่อที่แตกต่างกัน

2. ชื่อโดยทั่วไปจะแสดงความหมายของข้อมูลที่จะถูกจัดเก็บไว้ในคอลัมน์

3. ลำดับของคอลัมน์ไม่มีความสำคัญ


4) การใช้งานของโดเมน
ในแบบจำลองข้อมูลเชิงสัมพันธ์ ไม่มีการใช้ลิงค์ (Link) หรือพอยน์เตอร์ (Pointer)
เพื่อแสดงความสัมพันธ์ระหว่างข้อมูลในเชิงตรรกะ การหาความสัมพันธ์ในแบบจำลองข้อมูลเชิงสัมพันธ์  จึงทำโดยการเปรียบเทียบค่าของแอตทริบิวต์ที่ใช้ร่วมกัน  โดยที่คู่ของค่าเหล่านี้จะมีความหมายก็ต่อเมื่อเป็นค่าที่มาจากโดเมนเดียวกันดังแสดงในภาพข้างล่าง

domain usage


5) ดีกรี (Degree) และคาร์ดินัลลิตี (Cardinality) ของรีเลชัน

ดีกรีของรีเลชัน หมายถึง จำนวนของแอตทริบิวต์ (คอลัมน์) ในรีเลชันนั้น ส่วนคาร์ดินัลลิตีของรีเลชัน หมายถึง จำนวนของทัปเพิล (แถว) ในรีเลชันนั้น   ดังตัวอย่างที่แสดงไว้ในภาพข้างล่าง

degree


6) นิยามของรีเลชัน (Relation Definition)

รีเลชัน คือ สับเซตของคาร์ทีเซียนโปรดักต์ (Cartesian Product) ของโดเมน โดยกำหนดให เซต D1, D2,…Dn ซึ่งไม่มีความจำเป็นต้องเป็นโดเมนที่ต่างกัน  R  คือรีเลชันของเซตทั้ง n เซต          รีเลชัน จะประกอบด้วยเซตของทัปเพิล     โดยที่แต่ละทัปเพิลจะมีองค์ประกอบแรกจาก D1 องค์ประกอบที่สองจาก D2  ต่อเนื่องไปเรื่อย ๆ ตามลำดับ  เซตของ Di จะเรียกว่าเป็นโดเมนของ R โดยที่ค่าของ n  คือดีกรีของ R   และจำนวนทัปเพิลใน R  คือคาร์ดินัลลิตีของ R


7) รีเลชันนัลสกีมา (Relational Schema)
รีเลชันนัลสกีมา R  คือเซตของแอตทริบิวต์ R = {A1, A2, …, An}      และเขียนแทนด้วย R(A1, A2, …, An) หรือกล่าวอีกนัยหนึ่งก็คือนิยามของฐานข้อมูลซึ่งอธิบายถึงทุกคอลัมน์ในตาราง  และรีเลชันชิพที่มีในฐานข้อมูล


8) เงื่อนไข (Constraints)
เงื่อนไขในฐานข้อมูล หมายถึง  การกำหนดคุณสมบัติเพื่อจำกัดว่าข้อมูลชนิดใดบ้างที่จะสามารถนำมาจัดเก็บไว้ในแต่ละคอลัมน์ของฐานข้อมูลได้  แบบจำลองฐานข้อมูลเชิงสัมพันธ์    มีการกำหนดเงื่อนไขไว้โดยธรรมชาติน้อยมาก  แต่ผู้ใช้สามารถกำหนดเงื่อนไขต่าง ๆ ได้ตามที่ต้องการ โดยใช้คำสั่งของภาษาที่ใช้ในการนิยามข้อมูล DDL (Data Definition Language)

เงื่อนไขที่มีมาโดยธรรมชาติของรีเลชันในระบบฐานข้อมูลเชิงสัมพันธ์ คือ
1. ลำดับของคอลัมน์ไม่มีความสำคัญ
2. ลำดับของแถวหรือทัปเพิลไม่มีความสำคัญ
3. รีเลชันต้องประกอบด้วยทัปเพิลที่ไม่ซ้ำกัน
